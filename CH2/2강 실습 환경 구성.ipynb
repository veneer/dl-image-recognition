## 5. 이미지 데이터 이해하기

### 1) 이미지 데이터 구조
- 사람은 이미지를 보면 하나의 그림으로 인식하지만, 디지털 이미지는 무수히 많은 점들의 집합이다.
- 이 점들을 '픽셀'이라 하며, 한 픽셀의 색깔은 동일하다.

대체 텍스트



### 2) 흑백 이미지
- 각 픽셀에는 흑백의 밝기가 0 ~ 255의 값으로 표현되어 있다.

대체 텍스트



### 3) 컬러 이미지
- 빛을 구성하는 세 가지 색인 Red, Green, Blue를 이용해서 디지털 이미지를 표현한다.
- 세 가지 색의 배합에 따라 표현하는 색이 달라지며, 각 픽셀은 3개의 값을 가지고 있다.
- 세 가지 색은 각각 0에서 255 사이의 값을 갖고, 세 색깔이 모두 0이면 검은색 모두 255면 흰색이다.

대체 텍스트
